# **Fazendo webScraping com Beatifulsoup**

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get("https://quotes.toscrape.com/")
print(page.status_code)
```

```python
print(page.text)
```

```python
# lendo no beatifulsoup
dados_pagina = BeautifulSoup(page.text, "html.parser") # Pegando o tipo de objecto 
# formatando o html
# print(dados_pagina.prettify())
```

```python
txt = [] # onde vou armazenar os testos
frases = dados_pagina.find_all('span', class_="text")
for texto in frases:
    txt.append(texto.text)
    print(texto.text)
```

```python
# pegando os autores
author = []# onde vou armazenar os nomes dos autores
autores = dados_pagina.find_all('small', class_='author')
for nome_autores in autores:
    author.append(nome_autores.text)
    print(nome_autores.text)
```

# **Visualizando no pandas**

```python
import pandas as pd # importando a lib
```

```python
df = pd.DataFrame(zip(txt, author), columns = ['Texto', 'Autor'])
df
```

```python
# Salvando em formato csv
df.to_csv('Frases_motivacao.csv',index=None)
print('Dados salvo!')
```
